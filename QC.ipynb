```python
def State(c):
    simulator = Aer.get_backend('statevector_simulator')
    result = execute(c, simulator).result()
    c_state = result.get_statevector()
    return c_state

def getCounts(statevec,n):
 counts = {}
 for i, state in enumerate(statevec):
        binary_str = format(i, '0' + str(n) + 'b')
        probability = abs(state) ** 2
        counts[binary_str] = probability
     # Use counts to plot Histogram

 return counts
 
```

Quantum Teleportation 
1. Alice has qubits 0 and 1 while bob has qubit 2
2. qubit 1 and 2 are entangled in a bell state, and qubit 0 is in a random state
3. Then we measure qubit 0 and 1 and based on the results of classical registers, if q0 == 1, then Bob performs X on his qubit, and  q1 == 1 then he performs a z, none when q1 and q0 == 0 and both when q1=q0 == 1

```python
tc.cx(0,1)
tc.h(0)
State(tc).draw('latex')
```

```python
plot_bloch_multivector(State(tc))
```

```python
plot_histogram(getCounts(State(tc),3))
```

```python


tc.measure(0,0)
tc.measure(1,1)
tc.x(2).c_if(0,1)
tc.z(2).c_if(1,1)
tc.draw('mpl')
```

```python
plot_bloch_multivector(State(tc))
```
